```python
import networkx as nx
```

```python
G = nx.karate_club_graph()
print("Node Degree")

G = nx.convert_node_labels_to_integers(G, first_label=1)
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
# Compute the average shortest path length
short_part = nx.average_shortest_path_length(G)
short_part
```

```python
# Compute the average clustering coefficient
clust_part = nx.average_clustering(G)
clust_part
```

```python
# Compute the radius
radius_part = nx.radius(G)
radius_part
```

```python
# Compute the diameter
diameter_part = nx.diameter(G)
diameter_part
```

```python
# Center nodes
center_part = nx.center(G)
center_part
```

```python
# Periphery nodes
periphery_part = nx.periphery(G)
periphery_part
```

```python
# compute exentricity
exentricity_part = nx.eccentricity(G)
exentricity_part
```
